```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'pictures/t1/brown-eyes.jpg'
image_brown_eyes = cv2.imread(image_path)
image_path = 'pictures/t1/ronaldo.jpg'
image_ronaldo = cv2.imread(image_path)
```

```python
gray = cv2.cvtColor(image_brown_eyes, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image_ronaldo, cv2.COLOR_BGR2GRAY)
```

```python
#detection of the eyes
circles = cv2.HoughCircles(gray, method=cv2.HOUGH_GRADIENT, dp=1, minDist=20, param1=90, param2=40, minRadius=20, maxRadius=50)
image_copy = image_brown_eyes.copy()
if circles is not None:
    circles = np.uint16(np.around(circles))
    for circle in circles[0, :]:
        center = (circle[0], circle[1])
        radius = circle[2]
        cv2.circle(image_copy, center, radius, (0, 255, 0), 3)
image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#detection of the ballon d'Or's
circles = cv2.HoughCircles(gray2, method=cv2.HOUGH_GRADIENT, dp=1, minDist=35, param1=350, param2=45, minRadius=100, maxRadius=150)
image_copy = image_ronaldo.copy()
if circles is not None:
    circles = np.uint16(np.around(circles))
    for circle in circles[0, :]:
        center = (circle[0], circle[1])
        radius = circle[2]
        cv2.circle(image_copy, center, radius, (0, 255, 0), 5)
image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```
